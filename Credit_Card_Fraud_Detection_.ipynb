Importing the dependies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
 # chcecking the information dataset
 credit_card_data.info()
```

```python
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced


```python
ligit=credit_card_data[credit_card_data.Class == 0]
fraud=credit_card_data[credit_card_data.Class == 1]
```

```python
print(ligit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
ligit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compreing the values for the both transaction
credit_card_data.groupby('Class').mean()
```

Under sampling


build a sample dataset contaning simlar of normal transaction and fraud transcation

```python
ligit_sample=ligit.sample(n=492)
```

```python
new_dataset=pd.concat([ligit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#evaluation
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy of the training dataset",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy of the test dataset",test_data_accuracy)
```
